```python
df.quality.value_counts().plot(kind='bar')
```

```python
CCalidad=df.quality.value_counts()
```

```python
sb.set_style('darkgrid')
sb.barplot(x=CCalidad.index,y=CCalidad.values,color='red')
plt.xlabel('Calidad')
```

```python
columnas=df.select_dtypes(include="float64").columns
plt.figure(figsize=(14,len(columnas)*3))
for idx,c in enumerate(columnas,1):
  plt.subplot(len(columnas),2,idx)
  sb.histplot(df[c],kde=True)
  plt.title(c)
plt.tight_layout()

```

```python
plt.figure(figsize=(10,8))
sb.swarmplot(data=df,x='quality',y='alcohol',palette='viridis')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.title('Alcohol vs Quality')
```

```python
plt.figure(figsize=(30,28))
sb.pairplot(df,hue='quality')
plt.title('Pairplot of Wine Quality Dataset')
```

```python
plt.figure(figsize=(10,8))
sb.violinplot(data=df,x='quality',y='alcohol',palette='winter')
plt.title('Quality vs Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.title('Violin Plot of Alcohol vs Quality')
```

```python
sb.boxplot(data=df,x='quality',y='alcohol',palette='summer')
plt.title('Quality vs Alcohol')
```

```python
plt.figure(figsize=(15,10))
sb.heatmap(df.corr(),annot=True,fmt='.2f',cmap='Purples',linewidths=2)
plt.title('Correlation Heatmap')
```
