```python
ptip=(tips.tip/tips.total_bill)*100
```

```python
ptip=ptip.round(2)
```

```python
ptip
```

```python
tips['ptip']=ptip
```

```python
tips
```

```python
sb.relplot(data=tips,x='total_bill',y='ptip')
```

```python
sb.displot(tips.ptip,kde=True)
```

```python
st.skew(tips.ptip)
```

```python
st.kurtosis(tips.ptip)
```
