```python
nombres=['p'+str(i+1) for i in range(1000)]
calificaciones=list(range(5,11))
```

```python
nombres
```

```python
calificaciones
```

```python
np.random.choice(calificaciones)
```

```python
index=['m1','m2','m3','m4','m5','m6']
```

```python
data=pd.DataFrame()
for nombre in nombres:
  s=pd.Series(np.random.choice(calificaciones,6),index=index)
  data[nombre]=s
```

```python
data
```

```python
data.T.m6.mean()
```

```python
data.describe()
```

```python
st.mode(data.T)
```

```python
data.T.describe()
```
